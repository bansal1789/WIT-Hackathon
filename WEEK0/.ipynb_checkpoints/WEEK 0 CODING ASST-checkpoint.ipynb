```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

df_meal=pd.read_csv("meal_info.csv")
df_meal.head()


df_center=pd.read_csv("fulfilment_center_info.csv")
df_center.head()


df_food=pd.read_csv("train.csv")
df_food.head()

comb_data=pd.merge(df_food,df_center,on='center_id',how='inner')
df=pd.merge(comb_data,df_meal,on='meal_id',how='inner')
df.head()


table = pd.pivot_table(data=df,index='category',values='num_orders',aggfunc=np.sum)

'''can aggregate the data and summarise it efficiently in a form of a spreadsheet and data can be accessed quickly
Here index means based on which column/columns you want to sort or summarise and aggregate
Values means out of all columns in df which are required
aggfunc either sum/avg of the values are stored'''

dfr=table.reset_index()

label=dfr['category']
label
quantity=dfr['num_orders']
quantity
index = np.arange(len(label))
plt.bar(index, quantity,align='center',alpha=0.5)

plt.xticks(index,label,fontsize=5,rotation=70) 
plt.xlabel('Food items',fontsize=5)
plt.ylabel('Quantity sold',fontsize=5)
plt.title('Most popular food')
plt.savefig('./foodbar.png')
plt.show()


df['revenue'] = df['checkout_price'] * df['num_orders']
df['month']=df['week']//4


month=df['month'].unique()
month_order=pd.pivot_table(data=df,index='month',values='revenue',aggfunc=np.sum)


week=df['week'].unique()
week_order= pd.pivot_table(data=df,index='week',values='revenue',aggfunc=np.sum)

graph,(ax1,ax2)=plt.subplots(1,2,figsize=(20,5))
graph.suptitle('Horizontally stacked subplots')
ax1.plot(week,week_order)
ax1.set(xlabel='week',ylabel='revenue')
ax1.set_title('Weekly income')
#ax1.show()

ax2.plot(month,month_order)
ax2.set(xlabel='month',ylabel='revenue')
ax2.set_title('Monthly income')
#ax2.show()
plt.savefig('./subplots.png')
```
